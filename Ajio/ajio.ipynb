```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
s = Service("chromedriver.exe")
driver = webdriver.Chrome(service = s)

driver.get('https://www.ajio.com/s/men-watches-3991-40341')
driver.implicitly_wait(20)

old_height = driver.execute_script('return document.body.scrollHeight')

counter = 1
while True:
    try:
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
        time.sleep(3)
        new_height = driver.execute_script('return document.body.scrollHeight')

        if new_height == old_height:
            break
        
        print(old_height)
        print(new_height)
        print()
        old_height = new_height
    except:
        pass
    
    counter += 1

html = driver.page_source

with open('ajio.html','w',encoding='utf-8') as f:
    f.write(html)
```

```python
with open('ajio.html','r',encoding='utf-8') as f:
    html = f.read()

soup = BeautifulSoup(html,'lxml')
```

```python
container = soup.find_all('div',class_='contentHolder')
print(len(container))
```

```python
brand = []
name = []
rating = []
price = []
for i in container:
    # For Brand
    try:
        brand_name = i.find('div',class_='brand').text
        brand.append(brand_name)
    except:
        brand.append('-')
    
    # For Name
    try:
        watch_name = i.find('div',class_='nameCls').text
        name.append(watch_name)
    except:
        name.append('-')
    
    # For Rating
    try:
        watch_rating = i.find('p',class_='_1N0OO').text
        rating.append(watch_rating)
    except:
        rating.append('-')
    
    # For Price
    try:
        watch_price = i.find('div', attrs={'class':''}).text
        price.append(watch_price)
    except:
        price.append('-')
    
df = pd.DataFrame({'Brand':brand,'Name':name,'Rating':rating,'Price':price})
df.shape
```
