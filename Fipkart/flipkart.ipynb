```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup as bs
import pandas as pd

# Initialize web driver
s = Service("chromedriver.exe")
driver = webdriver.Chrome(service=s)

# Scraping
page_move = True
page_no = 1
html_content = ""
while page_move:
    print(page_no)
    if page_no == 42:
        page_move = False
        break
    else:
        driver.get("https://www.flipkart.com/search?q=smartphones&page={}".format(page_no))
        time.sleep(5)
        html_content += driver.page_source
        page_no += 1


# Save HTML content to a file
with open("flipkart_smartphones.html", "w", encoding="utf-8") as file:
    file.write(html_content)
```

```python
# Parsing offline HTML file using BeautifulSoup
with open("flipkart_smartphones.html", "r", encoding="utf-8") as file:
    soup = bs(file, 'html.parser')
```

```python
title = []
rating = []
storage = []
display = []
camera = []
battery = []
processor = []
price = []

pages = soup.find_all("div", attrs={"class": "DOjaWF gdgoEp"})
for page in pages:
    part = page.find_all("div", attrs={"class": "cPHDOP col-12-12"})
    for p in part:
        # For Title
        try:
            title_values = p.find('div', class_="KzDlHZ").text
            title.append(title_values)
        except:
            title.append('No Data')
        
        # For Rating
        try:
            rating_values = p.find('div', class_="XQDdHH").text
            rating.append(rating_values)
        except:
            rating.append('No Data')
        
        # For Storage
        try:
            storage_elem = p.find('ul', class_="G4BRas").find_all('li')
            storage_val = storage_elem[0].text if storage_elem else 'No Data'
            storage.append(storage_val)
        except:
            storage.append('No Data')
        
        # For Display
        try:
            display_val = storage_elem[1].text if len(storage_elem) > 1 else 'No Data'
            display.append(display_val)
        except:
            display.append('No Data')
        
        # For Camera
        try:
            camera_val = storage_elem[2].text if len(storage_elem) > 2 else 'No Data'
            camera.append(camera_val)
        except:
            camera.append('No Data')
        
        # For Battery
        try:
            battery_val = storage_elem[3].text if len(storage_elem) > 3 else 'No Data'
            battery.append(battery_val)
        except:
            battery.append('No Data')
        
        # For Processor
        try:
            processor_val = storage_elem[4].text if len(storage_elem) >= 4 else 'No Data'
            processor.append(processor_val)
        except:
            processor.append('No Data')
        
        # For Price
        try:
            price_val = p.find('div', class_="Nx9bqj _4b5DiR").text
            price.append(price_val)
        except:
            price.append('No Data')

df = pd.DataFrame({'Title': title, 'Rating': rating, 'Storage': storage, 'Display': display, 'Camera': camera, 'Battery': battery, 'Processor': processor, 'Price': price})
```
